```python
def plot_background(figsize=(12,12)):
    water = geopandas.read_file("data/water-shp")
    city = geopandas.read_file("data/madison-shp")
    colored_routes = geopandas.read_file("Madison-Metro-Sim/data/andy_files/colored_routes-shp")

    # ax = colored_routes.plot(figsize=figsize, color=colored_routes["route_colo"])
    ax = water.plot(figsize=figsize, color="lightblue")
    city.plot(color="0.85", ax=ax)

    # ax.set_axis_off()
    return ax

def plot_route(route_num=None): # TODO, add bus stop
    ax = plot_background()

    if route_num != None:
        wanted = colored_routes[colored_routes["route_shor"] == route_num]
        wanted.plot(color=wanted["route_colo"], ax=ax)

    else:
        colored_routes.plot(color=colored_routes["route_colo"], ax=ax)

    return ax

plot_route()
# colored_routes[colored_routes["route_shor"] == 80].head()
```
