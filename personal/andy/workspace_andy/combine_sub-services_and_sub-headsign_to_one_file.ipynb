```python
filepath = "../../../data/mmt_gtfs/routes"

import pandas as pd
import os
```

```python
os.listdir(filepath)
```

```python
df = pd.DataFrame()

for service in os.listdir(filepath):
    if service[:2] != "92":
        continue
        print(service)
    for route in os.listdir(os.path.join(filepath, service)):
        # print(service, route[5:route.index("_")])
        route_num = route[5:route.index("_")]
        route_df = pd.read_csv(os.path.join(filepath, service, route))
        print(len(route_df))
        route_df.insert(loc=0, column="route_num",
                        value=int(route_num))
        route_df["service_id"] = service
        df = df.append(route_df, ignore_index=True)


```

```python
filename = os.path.join(filepath, "all_routes.csv")
df.to_csv(path_or_buf=filename, index=False)
```
